Let D be a database on B,U⊆B an itemset, and minsupp∈Z a natural number. We call U frequent in database D if Supp(U,D)≥minsupp.

Nós também podemos definir:
Seja F o conjunto de todos os frequent itemset. Um conjunto de itens U pertence ao conjunto F se, e somente se, o suporte de U no banco de dados D for maior ou igual ao suporte mínimo minsupp.
U∈F⟺Supp(U,D)≥minsup
A partir disso, nós podemos definir o frequent itemset usando o suporte relativo (ou proporcional) denotado por minsupp%:
0≤minsup%≤1
Assim, um itemset também é frequente se, e somente se, o suporte relativo de U no banco de dados D for maior ou igual ao minsupp relativo.
Supp%(U,D)≥minsupp%
Disso, podemos afirmar que uma amostra s de itens frequentes extraída de F (denotada por F ̃_s) também obedece ao princípio Apriori. Assim, o princípio Apriori pode ser aplicado nas amostras do particionamento e do processamento paralelo.
