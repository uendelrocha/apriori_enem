```python
# Variáveis categóricas que contém texto
cat_texto = [
  'NO_MUNICIPIO_ESC', 'NO_MUNICIPIO_PROVA',
  'Q001', 'Q002', 'Q003', 'Q004', 'Q005', 'Q006', 'Q007', 'Q008', 'Q009', 'Q010', 
  'Q011', 'Q012', 'Q013', 'Q014', 'Q015', 'Q016', 'Q017', 'Q018', 'Q019', 'Q020', 
  'Q021', 'Q022', 'Q023', 'Q024', 'Q025', 
  'SG_UF_ESC', 'SG_UF_PROVA', 
  'TP_LINGUA', 
  'TP_SEXO', 
]

# Variáveis categóricas que contém números
cat_numero = [
  'CO_MUNICIPIO_ESC', 'CO_MUNICIPIO_PROVA', 
  'CO_PROVA_CH', 'CO_PROVA_CN', 'CO_PROVA_LC', 'CO_PROVA_MT', 
  'CO_UF_ESC', 'CO_UF_PROVA',
  'IN_TREINEIRO', 
  'NU_INSCRICAO', 'NU_ANO',
  'TP_ANO_CONCLUIU', 'TP_COR_RACA', 'TP_DEPENDENCIA_ADM_ESC', 'TP_ENSINO', 'TP_ESCOLA', 'TP_ESTADO_CIVIL', 
  'TP_FAIXA_ETARIA', 'TP_LOCALIZACAO_ESC', 'TP_NACIONALIDADE', 
  'TP_PRESENCA_CH', 'TP_PRESENCA_CN', 'TP_PRESENCA_LC', 'TP_PRESENCA_MT',
  'TP_SIT_FUNC_ESC', 'TP_ST_CONCLUSAO', 'TP_STATUS_REDACAO',
]

cat_vetor = [
  'TX_GABARITO_CH', 'TX_GABARITO_CN', 'TX_GABARITO_LC', 'TX_GABARITO_MT', 
  'TX_RESPOSTAS_CH', 'TX_RESPOSTAS_CN', 'TX_RESPOSTAS_LC', 'TX_RESPOSTAS_MT'
]

categoricas = cat_texto + cat_numero + cat_vetor

continuas = [
  'NU_NOTA_CH', 'NU_NOTA_CN', 
  'NU_NOTA_COMP1', 'NU_NOTA_COMP2', 'NU_NOTA_COMP3', 'NU_NOTA_COMP4', 'NU_NOTA_COMP5', 
  'NU_NOTA_LC', 'NU_NOTA_MT', 'NU_NOTA_REDACAO',
]
```

```python
# Colunas que serão removidas por não interessarem para a análise
colunas_rem = []
#colunas_rem =  ['NU_INSCRICAO']
colunas_rem =  ['NU_ANO']
colunas_rem += ['TX_RESPOSTAS_CN', 'TX_RESPOSTAS_CH', 'TX_RESPOSTAS_LC', 'TX_RESPOSTAS_MT']
colunas_rem += ['TX_GABARITO_CN', 'TX_GABARITO_CH', 'TX_GABARITO_LC', 'TX_GABARITO_MT']
colunas_rem += ['CO_PROVA_CN', 'CO_PROVA_CH', 'CO_PROVA_LC', 'CO_PROVA_MT']
#colunas_rem += ['TP_ENSINO', 'TP_DEPENDENCIA_ADM_ESC', 'TP_LOCALIZACAO_ESC', 'TP_SIT_FUNC_ESC']
colunas_rem += ['CO_MUNICIPIO_ESC', 'NO_MUNICIPIO_ESC', 'CO_MUNICIPIO_PROVA', 'NO_MUNICIPIO_PROVA']
colunas_rem += ['CO_UF_ESC']#, 'SG_UF_ESC']
colunas_rem += ['CO_UF_PROVA']#, 'SG_UF_PROVA']
colunas_rem += ['TP_PRESENCA_CN', 'TP_PRESENCA_CH', 'TP_PRESENCA_LC', 'TP_PRESENCA_MT']
colunas_rem += ['TP_STATUS_REDACAO']
```

```python
# Remover registros de inscritos que faltaram ou que tiveram problemas com a redação
df_presentes = df_enem.query('TP_PRESENCA_CN == 1 & TP_PRESENCA_CH == 1 & TP_PRESENCA_LC == 1 & TP_PRESENCA_MT == 1').copy(deep=True)
df_presentes = df_enem.query('TP_STATUS_REDACAO == 1').copy(deep=True)
df_presentes.head(5)
```

```python
# Remove as colunas que não interessam
df_presentes = df_presentes.drop(columns=colunas_rem)
df_presentes.reset_index(drop=True, inplace=True)
df_presentes.head(5)
```

```python
# Converte as colunas categóricas para object
cat_in_df = [c for c in categoricas if c in df_presentes.columns]
for c in cat_in_df:
  df_presentes[c] = df_presentes[c].astype(object)
```

```python
df_presentes.info(show_counts=True)
```

```python
# Alterar para nan as variáveis não informadas nos registros
# Isso é necessário para ajustar os valores ao problema
# Um dado classificado como não informado, não sei ou não respondeu é um dado ausente
NAO_INFORMADOS_0 = ['TP_ESTADO_CIVIL', 'TP_COR_RACA', 'TP_NACIONALIDADE', 'TP_ANO_CONCLUIU']
NAO_RESPONDEU_1 = ['TP_ESCOLA']
NAO_SEI_H = ['Q001', 'Q002']
NAO_SEI_F = ['Q003', 'Q004']
NI_COLS = NAO_INFORMADOS_0 + NAO_RESPONDEU_1 + NAO_SEI_H + NAO_SEI_F
NI_COLS = [x for x in NI_COLS if x in df_presentes.columns]
for index, row in df_presentes.iterrows():
  for col_name in NI_COLS:
    if col_name in NAO_INFORMADOS_0:
      if row[col_name] == 0:
        df_presentes.at[index, col_name] = pd.NA
    elif col_name in NAO_RESPONDEU_1:
      if row[col_name] == 1:
        df_presentes.at[index, col_name] = pd.NA
    elif col_name in NAO_SEI_H:
      if row[col_name] == 'H':
        df_presentes.at[index, col_name] = pd.NA
    elif col_name in NAO_SEI_F:
      if row[col_name] == 'F':
        df_presentes.at[index, col_name] = pd.NA
```
