```python
import pandas as pd
```

```python
ENCODING = 'windows_1252'
ANO_ENEM = 2022
ANO_ENEM = str(ANO_ENEM)
PATH_ENEM = r'Y:\DATASOURCES\enem'
PATH_MD = r'{}\microdados_enem_{}\DADOS'.format(PATH_ENEM, ANO_ENEM)
CSV_MD = f'{PATH_MD}\\MICRODADOS_ENEM_{ANO_ENEM}.csv'
PARQUET_MD = f'{PATH_MD}\\MICRODADOS_ENEM_{ANO_ENEM}.parquet'
PARQUET_SAN = f'{PATH_MD}\\MICRODADOS_ENEM_{ANO_ENEM}_SANITIZED.parquet'
FEATHER_SAN = f'{PATH_MD}\\MICRODADOS_ENEM_{ANO_ENEM}_SANITIZED.feather'
CSV_SAN = f'{PATH_MD}\\MICRODADOS_ENEM_{ANO_ENEM}_SANITIZED.csv'
```

```python
print(PARQUET_MD)
```

```python
df_enem = pd.read_parquet(PARQUET_MD)
```

```python
df_enem.columns
```

```python
df_enem.info(show_counts=True)
```

```python
# Remove colunas que não interessam
colunas_rem =  ['NU_INSCRICAO']
colunas_rem += ['TX_RESPOSTAS_CN', 'TX_RESPOSTAS_CH', 'TX_RESPOSTAS_LC', 'TX_RESPOSTAS_MT']
colunas_rem += ['TX_GABARITO_CN', 'TX_GABARITO_CH', 'TX_GABARITO_LC', 'TX_GABARITO_MT']
colunas_rem += ['CO_PROVA_CN', 'CO_PROVA_CH', 'CO_PROVA_LC', 'CO_PROVA_MT']
#colunas_rem += ['TP_ENSINO', 'TP_DEPENDENCIA_ADM_ESC', 'TP_LOCALIZACAO_ESC', 'TP_SIT_FUNC_ESC']
colunas_rem += ['CO_MUNICIPIO_ESC', 'NO_MUNICIPIO_ESC', 'CO_MUNICIPIO_PROVA', 'NO_MUNICIPIO_PROVA']
colunas_rem += ['CO_UF_ESC', 'SG_UF_ESC', 'CO_UF_PROVA', 'SG_UF_PROVA']
colunas_rem += ['TP_PRESENCA_CN', 'TP_PRESENCA_CH', 'TP_PRESENCA_LC', 'TP_PRESENCA_MT']
colunas_rem
```

```python
# Remover registros de inscritos que faltaram
df_presentes = df_enem.query('TP_PRESENCA_CN == 1 & TP_PRESENCA_CH == 1 & TP_PRESENCA_LC == 1 & TP_PRESENCA_MT == 1').copy(deep=True)
df_presentes.head(5)
```

```python
df_presentes = df_presentes.drop(columns=colunas_rem)
df_presentes.reset_index(drop=True, inplace=True)
df_presentes.head(5)
```

```python
df_presentes.info(show_counts=True)
```

```python
categoricas = [
  'NU_ANO',
  'TP_DEPENDENCIA_ADM_ESC' , 'TP_LOCALIZACAO_ESC', 'TP_LOCALIZACAO_ESC', 'TP_SIT_FUNC_ESC', 'TP_ENSINO',
  'TP_FAIXA_ETARIA', 'TP_ESTADO_CIVIL', 'TP_COR_RACA', 'TP_NACIONALIDADE',
  'TP_ST_CONCLUSAO', 'TP_ANO_CONCLUIU', 'TP_ESCOLA', 'IN_TREINEIRO',
  'TP_LINGUA', 'TP_STATUS_REDACAO', 'Q001', 'Q002', 'Q003', 'Q004',
  'Q005', 'Q006', 'Q007', 'Q008', 'Q009', 'Q010', 'Q011', 'Q012',
  'Q013', 'Q014', 'Q015', 'Q016', 'Q017', 'Q018', 'Q019', 'Q020',
  'Q021', 'Q022', 'Q023', 'Q024', 'Q025',
]

df = df_presentes.copy(deep=True)
for c in categoricas:
  df[c] = df[c].astype(str)
```

```python
df.info()
```

```python
df_presentes = df.copy(deep=True)
```

```python
df_presentes.to_csv(CSV_SAN, sep=';', encoding=ENCODING)
```

```python
df_presentes.to_parquet(path=PARQUET_SAN, engine='auto', compression='brotli')
```
